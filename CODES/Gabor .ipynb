```python
import matplotlib.pyplot as plt
import numpy as np
import os
import cv2
import time
```

```python
def load_images():
    dir = os.path.join(os.sep,'home','mohit','Desktop','dataset')
    images = {}
    for k in range(1,113):
        for i in ['1','2']:
            for j in ['1','2','3','4','5','6','7','8']:
                try:
                    images['c'+str(k)+'_'+i+'_'+j] = cv2.imread(os.path.join(dir,'Cheque'+str(k),i,i+'_'+j+'.tif'))
                    images['c'+str(k)+'_'+i+'_'+j] = cv2.cvtColor(images['c'+str(k)+'_'+i+'_'+j], cv2.COLOR_BGR2YCrCb)
                except:
                    break
    return images
```

```python
def n_size(images):
    size = {} #no. of ink pixels .. considering all non white pixels(<th) as ink pixels
    for k in range(1,113):
        for i in ['1','2']:
            for j in ['1','2','3','4','5','6','7','8']:
                count = 0
                try:
                    image = images['c'+str(k)+'_'+i+'_'+j][:,:,0]
                    for a in range(0,image.shape[0]):
                        for b in range(0,image.shape[1]):
                            if image[a][b] != 255:
                                count+=1
                    size['c'+str(k)+'_'+i+'_'+j] = count
                   # print('c'+str(k)+'_'+i+'_'+j,count)
                except:
                    #print('..')
                    break  
    return size
```

```python
def features(img1,img2,s):
    fv1,fv2 = 0,0
    for x in range(0, len(img1)):
        for y in range(0, len(img1[0])):
            if(img1[x,y]!=255):
                fv1+=img2[x,y]
    fv1/=s        
    for x in range(0, len(img1)):
        for y in range(0, len(img1[0])):
            if(img1[x,y]!=255):    
                fv2+=(img2[x,y]-fv1)**2
    fv2/=(s-1)
    fv2=fv2**0.5
    return fv1,fv2
```

```python
def calc_gabor(images,dim,sizes): 
    f = {}
    g_kernel = []
    
    ksize = (9,9)
    sigma = [1,1.5,2,2.5,3]
    theta = [0,np.pi/4,np.pi/2,3*np.pi/4]
    lamda = [2,2.5,3,3.5,4]
    gamma = 0.5
    psi = np.pi
    
    #need to set kernel size, sigma and lambda 
    for i in sigma:
        for j in lamda:
            for k in theta:
                g_kernel.append(cv2.getGaborKernel(ksize, i, k, j, gamma, psi, ktype=cv2.CV_32F))
    
    for k in range(1,113):
        print(k)
        for i in ['1','2']:
            for j in ['1','2','3','4','5','6','7','8']:
                try:
                    s = sizes['c'+str(k)+'_'+i+'_'+j]
                    h = []
                    for g in g_kernel:
                        gabor = cv2.filter2D(images['c'+str(k)+'_'+i+'_'+j][:,:,dim], cv2.CV_32FC1, g)
                        h.append(features(images['c'+str(k)+'_'+i+'_'+j][:,:,0],gabor,s))
                        
                    f['c'+str(k)+'_'+i+'_'+j] = np.array(h).ravel()
                except:
                    break
                   
    return f
```

```python
images= load_images()
```

```python
sizes = n_size(images)
```

```python
hist ={}
hist = calc_gabor(images,0,sizes)
```

```python
hist1 = {}
hist1 = calc_gabor(images,1,sizes)
```

```python
hist2 = {}
hist2 = calc_gabor(images,2,sizes)
```
